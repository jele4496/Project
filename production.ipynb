```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
data = pd.read_csv(data_source)
```

```python
#groups dataframe "data" by "Year", select the column from dataframe, and calculate the sum of column's value
annual_production = data.groupby('Year')['Production'].sum()
```

```python
#generate line chart
plt.figure(figsize = (10, 6))
plt.plot(annual_production.index, annual_production.values, marker = 'o', color = 'blue') #marker used to make a dot on the line chart
plt.title('Annual Honey Production') #title
plt.xlabel('Year') #xlabel
plt.ylabel('Production')#ylabel
plt.show() #print chart
```
