```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
data = pd.read_csv(data_source)
```

```python
annual_production = data.groupby('Year')['Production'].sum()
```

```python
plt.figure(figsize=(10, 6))
annual_production.plot(kind='bar', color='green')
plt.xlabel('Year')
plt.ylabel('Total Honey Production (in thousands of pounds)')
plt.title('Total Honey Production Each Year (2019-2023)')
plt.xticks(rotation=0)
plt.tight_layout()
```
