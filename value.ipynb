```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
data = pd.read_csv(data_source)
```

```python
#groups dataframe "data" by "Year", select the column from dataframe, and calculate the sum of column's value
annual_value = data.groupby('Year')['Value of production'].sum()
```

```python
#generate line chart
plt.figure(figsize = (10, 6))
plt.plot(annual_value.index, annual_value.values, marker = 'o', color = 'green') #marker used to make a dot on line chart
plt.title('Annual Value of production') #title
plt.xlabel('Year') #xlabel
plt.ylabel('Value of production') #ylabel
plt.show() #print chart
```
