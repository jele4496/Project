```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
data = pd.read_csv(data_source)
```

```python
#groups dataframe "data" by "Year", select the column from dataframe, and calculate the sum of column's value
annual_yield = data.groupby('Year')['Yield per colony'].sum()
```

```python
#generate line chart
plt.figure(figsize = (10, 6))
plt.plot(annual_yield.index, annual_yield.values, marker = 'o', color = 'yellow') #marker used to make a dot on line chart
plt.title('Annual Yield per Colony') #title
plt.xlabel('Year') #xlabel
plt.ylabel('Yield per Colony') #ylabel
plt.show() #print chart
```
