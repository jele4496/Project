```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
data = pd.read_csv(data_source)
```

```python
annual_yield = data.groupby('Year')['Yield per colony'].sum()
```

```python
plt.figure(figsize=(10, 6))
annual_yield.plot(kind='bar', color='blue')
plt.xlabel('Year')
plt.ylabel('Average Yield per Colony (in pounds)')
plt.title('Average Yield per Colony Each Year (2019-2023)')
plt.xticks(rotation=0)
plt.tight_layout()
```
