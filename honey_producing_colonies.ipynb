```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
```

```python
data = pd.read_csv(data_source)
```

```python
#groups dataframe "data" by "Year", select the column from dataframe, and calculate the sum of column's value
annual_colonies = data.groupby('Year')['Honey producing colonies'].sum()
```

```python
#generate the line chart
plt.figure(figsize = (10, 6))
plt.plot(annual_colonies.index, annual_colonies.values, marker = 'o', color = 'black') #use marker to put the dot on the line chart
plt.title('Annual Honey Producing Colonies') #set title
plt.xlabel('Year') #set xlabel
plt.ylabel('Honey Producing Colonies') #set ylabel
plt.show() #print chart
```
