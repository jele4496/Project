```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
data_source = 'honey_2019_2023.csv'
```

```python
data = pd.read_csv(data_source)
```

```python
#check Nan Value in column and see how many of them
annual_colonies = data.groupby('Year')['Honey producing colonies'].sum()
```

```python
#drop column that is Nan value
data_cleaned = data.dropna(subset=['State'])
```

```python
plt.figure(figsize=(10, 6))
annual_colonies.plot(kind='bar', color='goldenrod')
plt.xlabel('Year')
plt.ylabel('Total Honey Producing Colonies (in thousands)')
plt.title('Total Honey Producing Colonies Each Year (2019-2023)')
plt.xticks(rotation=0)
plt.tight_layout()
```
